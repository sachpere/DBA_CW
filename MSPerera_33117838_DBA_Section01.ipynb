```r
install.packages("sqldf")
library(sqldf)

energy_data <- read.csv("Energy_dataset.csv")
head(energy_data)
```

```r
result <- sqldf("SELECT * FROM energy_data WHERE 'Overall.Height' IS NULL ")

head(result)
```

```r
result1 <- sqldf("SELECT * FROM energy_data")

print(result1)
```

```r
result2 <- sqldf("SELECT * FROM energy_data WHERE 'relative.compactness' > 0.7")

head(result2)
```

```r
library(dplyr)


filtered_data <- energy_data %>%
  filter(Relative.Compactness > 0.5)

print(filtered_data)
```

```r
selected_data <- energy_data %>%
  select(Relative.Compactness, Surface.Area, Heating.Load)

print(selected_data)
```

```r
sorted_data <- energy_data %>%
  arrange(desc(Overall.Height))

print(sorted_data)
```

```r
library(ggplot2)


scatter_plot <- ggplot(energy_data, aes(x = Surface.Area, y = Heating.Load)) +
                geom_point() +
                labs(x = "Surface Area", y = "Heating Load") +
                ggtitle("Scatter Plot of Surface Area vs. Heating Load")

print(scatter_plot)
```

```r
histogram <- ggplot(energy_data, aes(x = Heating.Load)) +
             geom_histogram(binwidth = 5, fill = "skyblue", color = "black") +
             labs(x = "Heating Load", y = "Frequency") +
             ggtitle("Histogram of Heating Load")


print(histogram)
```

```r
bar_plot <- ggplot(energy_data, aes(x = Orientation, y = Cooling.Load)) +
            geom_bar(stat = "summary", fun = "mean", fill = "lightgreen") +
            labs(x = "Orientation", y = "Average Cooling Load") +
            ggtitle("Average Cooling Load by Orientation")


print(bar_plot)
```
